In the example below, our ATM will test whether the user knows the account's secret PIN Number

```python
stored_pin = "0000"
user_pin = await input("Enter your ATM PIN password")
```

```python
if (stored_pin != user_pin):
    print("PIN verification failed")
else:
    print("PIN verified!")
```

You can even select through many outcomes with 'elif'

```python
a = 33
b = 33
if b > a :
    print("b is greater than a")
elif a == b :
    print("a and b are equal")
else :
    print("b is less than a")
```

Fill in the blanks in the code snippet below so that the ATM client cannot withdraw more than 500 dollars at a time, even if they have enough money

```python
balance = 1000 #dollars

withdrawl = int(await input(
    "How much would you like to withdraw?"))
```

```python
if withdrawl <= balance and withdrawl <= 500:
    balance = balance - withdrawl
    print("Your new balance is: " + str(balance))
elif withdrawl > 500:
    print("You may not withdraw more than 500 dollars at a time")
else:
    print("Not enough money!")
print("Goodbye!")
```
