in the following example, the function mario_gain_coin expects the parameter coin_value

```python
coins = 0
lives = 3

def mario_gain_coin(coin_value) :      #mario_gain_coin expects the parameter coin_value
    global coins
    global lives
    coins = coins + coin_value
    if (coins >= 100) :                 #mario should exchange 100 coins for 1 life automatically
        lives = lives + coins // 100
        coins = coins % 100
    print("coins = " + str(coins))
    print("lives = " + str(lives))
    
mario_gain_coin(1)        
mario_gain_coin(10)
mario_gain_coin(100)
mario_gain_coin(888)
```

You can use paramters to pass arguments (data) do a function, but what if you want the function to return a value back to you? For that you can use 'return'. The example below will tell you the next day of the week, but also give the value back to you in the return statement at the end.

```python
def next_day(day) :
    if (day == "monday") :
        day = "tuesday"
        print("Now it's " + day)
    elif (day == "tuesday") :
        day = "wednesday" 
        print("Now it's " + day)
    elif (day == "wednesday") :
        day = "thursday"
        print("Now it's " + day)
    elif (day == "thursday") :
        day = "friday"
        print("Now it's " + day)
    elif (day == "friday") :
        day = "saturday"
        print("Now it's " + day)
    elif (day == "saturday") :
        day = "sunday"
        print("Now it's " + day)
    elif (day == "sunday") :      #on sunday, loop back to monday
        day = "monday"
        print("Now it's " + day)
    return day                    #return the value of 'day' back to the function call
    
x = "thursday"
x = next_day(x)                   #call next_day, suppling the argument x, and returning back to x
```

Functions can also expect many parameters. The function triangle_type expects the lengths of the sides of a triangle and will report whether the triangle is right, obtuse, or acute

<div><img src=attachment:d05de872-0fa4-4138-af55-4f0e2a593621.png width = 600></div>

```python
def triangle_type(side1, side2, side3) :
    #figure out which sides are which
    max_side = max(side1, side2, side3)
    min_side = min(side1, side2, side3)
    middle_side = side1 + side2 + side3 - max_side - min_side
    
    if (max_side**2 > min_side**2 + middle_side**2):
        print("this triangle is obtuse")
    elif (max_side**2 < min_side**2 + middle_side**2):
        print("this triangle is acute")
    else :
        print("this triangle is right")
        
```
