Loops might repeat a specific number of times:

```python
factorial = 1  #this is a counter
operand = 4    #will will find the factorial of this number. 
               #Try picking a different integer and see what happens

for x in range(1, operand+1) :
    factorial = factorial * x
    
print(str(operand) + "! is " + str(factorial))
```
