Modulus (%) is a math operation you might not be familiar with. Here's an example:

```python
chocolate_squares = 16
people = 3

print("Everyone can have " + str(chocolate_squares // people) + " chocolates")
print("but there will be " + str(chocolate_squares % people) + " chocolate(s) left over")
```

The modulus operation computes the remainder of a division. In the example above, 16 divided by 3 was 5 with a remainder of 1. so:

16 // 3 = 5
16 % 3 = 1

Can you guess the results of the following modulus operations?

```python
print(20%3)
print(14%5)
print(14%3)
print(14%7)
print(14%14)
print(14%16)
```
